```python
#Function for 1a - Longest period of consecutive values above the mean
#TIMESERIES NOTE: 1-d list of values (index is time)
def longestconsecval(df):
    try:
        df.astype('float64', errors = 'raise') #convert all data to floats, if this doesn't work (meaning wrong datatype input) --> raise error
    except:
        print('Unexpected datatype! Only integers or floats are accepted for this function.') #wrong datatype input
    else:
        #NEED TO CHECK IF THE VALUES IN DF ARE NUMBERS
        avgval = np.nanmean(df) #ignores missing values
        length = len(df)  #num of rows
        iscurrentperiod = False    #boolean conditional to see if a period containing values fitting requirements is going on
        currentperiod = 0      #place to store ongoing period
        count = 0            #counter for loop
        maxperiod = 0         #where we will store the longest period (what we are looking for)
        while count < length:   #going through each row
            if np.isnan(df.iloc[count])[0]:    #if value is not there, increase the counter and start loop again (indexing at 0 bc isnan returns array)
                count +=1
                continue
            elif (df.iloc[count,0] > avgval) and iscurrentperiod == True:  #this value is above the average and an above average period is ongoing
                currentperiod += 1      #adds another time unit to the maximum period (another index containing an above avg value)
                if currentperiod > maxperiod:  #checks if current ongoing period is larger than max, if so becomes new max
                    maxperiod = currentperiod
            elif (df.iloc[count,0] > avgval) and iscurrentperiod == False:    #a period is not ongoing, but current value is above average
                iscurrentperiod = True         #a new period will start now
                currentperiod = 1            #starting a new period
            elif (df.iloc[count,0] <= avgval):            #there was an ongoing period, but current value is not above avg, end period
                iscurrentperiod = False
                currentperiod = 0
            count += 1            #incrementing at end of loop
        if maxperiod == 0:
            print('No consecutive values are above the mean of given values.')
        else:
            print('The longest period of consecutive values above the mean is ' + str(maxperiod) + '.')
```

```python
#Function for 1b - Time intervals between successive extreme events above the mean
    #we are outputting a datafram of time intervals
def successiveExtremeEventsAboveMean(df, threshold):
    try:
        df.astype('float64', errors = 'raise') #convert all data to floats, if this doesn't work (meaning wrong datatype input) --> raise error
        threshold = float(threshold)
    except:
        print('Unexpected datatype in time-series or threshold! Only integers or floats are accepted for this function.') #wrong datatype input
    else:
        avgval = np.nanmean(df) #ignores missing values
        if threshold <= avgval:          #checking that threshold is above average
            print('Threshold is below the average and needs to be above.')
        timeintervalsaboveavg = pd.DataFrame(columns = ['Time intervals between successive extreme events above the mean'])  #where to store the intervals
        length = len(df)  #num of rows
        count = 0
        startingtimestamp = ''           #where the time interval will start initially
        while count < length:
            if np.isnan(df.iloc[count])[0]: #if value is not there, increase the counter and start loop again (indexing at 0 bc isnan returns array)
                count +=1
                continue
            elif (df.iloc[count,0] > threshold):  #this value is above the threshold for extreme
                
                if startingtimestamp == '':           #this is the first value in the data above the threshold
                    startingtimestamp = count
                else:                                #there was a prev value above the threshold
                    interval = count - startingtimestamp             #relevant time interval
                    #add interval to dataframe
                    timeintervalsaboveavg = timeintervalsaboveavg.append({'Time intervals between successive extreme events above the mean': interval}, ignore_index=True)
                    startingtimestamp = count           #reset starting time interval
            elif (df.iloc[count,0] <= threshold):   #this value is below the threshold
                pass                              #nothing happens
                
            count += 1   #incrementing count at end of loop
        print(timeintervalsaboveavg)



       
```

```python
#Function for 1c - Time intervals between successive extreme events below the mean

    #assuming we are outputting an array of time intervals
def successiveExtremeEventsBelowMean(df, threshold):
    try:
        df.astype('float64', errors = 'raise') #convert all data to floats, if this doesn't work (meaning wrong datatype input) --> raise error
        threshold = float(threshold)
    except:
        print('Unexpected datatype in time-series or threshold! Only integers or floats are accepted for this function.') #wrong datatype input
    else:
        avgval = np.nanmean(df) #ignores missing values
        if threshold >= avgval:          #checking that threshold is below average
            print('Threshold is above the average and needs to be below.')
        timeintervalsbelowavg = pd.DataFrame(columns = ['Time intervals between successive extreme events below the mean'])  #where to store the intervals
        length = len(df)  #num of rows
        count = 0
        startingtimestamp = ''           #where the time interval will start initially
        while count < length:      #while list is still goin
            if np.isnan(df.iloc[count])[0]: #if value is not there, increase the counter and start loop again (indexing at 0 bc isnan returns array)
                count +=1
                continue
            elif (df.iloc[count,0] < threshold):  #this value is below the threshold for extreme
                
                if startingtimestamp == '':           #this is the first value in the data below the threshold
                    startingtimestamp = count
                else:                                #there was a prev value above the threshold
                    interval = count - startingtimestamp             #relevant time interval
                    #add interval to dataframe
                    timeintervalsbelowavg = timeintervalsbelowavg.append({'Time intervals between successive extreme events below the mean': interval}, ignore_index=True)
                    startingtimestamp = count           #reset starting time interval
            elif (df.iloc[count,0] >= threshold):   #this value is above the threshold
                pass                              #nothing happens
                
            count += 1   #incrementing count at end of loop
        print(timeintervalsbelowavg)



       
```

```python
df_covid = pd.read_csv('COVID-19 Nursing Home Data 01.15.2023.csv')
df_covid
```

```python
df_covid['Week Ending'] = pd.to_datetime(df_covid['Week Ending'])
df_covid
```

1. Generate a table showing the distribution of the number of nursing home facilities by state and county. (10 points)





```python
#Solution for Problem 2 Task 1
df_nursinghome_distrib = df_covid.copy()
df_nursinghome_distrib = df_nursinghome_distrib.groupby(['Provider State', 'County'], as_index = False)['Provider Name'].nunique()  #groups by state and county and counts unique nursing home facilities
df_nursinghome_distrib.rename(columns = {'Provider Name':'Nursing Home Count'}, inplace = True)
df_nursinghome_distrib
```

2. Compute the average weekly %occupancy by state for each year (20 Points)

```python
#Solution for Problem 2 Task 2
df_percoccupancy = df_covid.copy()
df_percoccupancy['Year'] = df_percoccupancy['Week Ending'].dt.year

df_percoccupancy = df_percoccupancy.dropna(subset = ['Total Number of Occupied Beds'])
df_percoccupancy = df_percoccupancy.dropna(subset = ['Number of All Beds'])

df_percoccupancy['% Occupancy'] = df_percoccupancy['Total Number of Occupied Beds']/df_percoccupancy['Number of All Beds'] *100
df_percoccupancygrouped = df_percoccupancy.groupby(['Year', 'Provider State'])
df_averageweeklyoccupancy = df_percoccupancygrouped['% Occupancy'].mean()
print(df_averageweeklyoccupancy)
```

3. List the top 5 nursing homes that faced maximum consecutive weeks of nursing staff shortage (20 points)

```python
df_nursingstaffshortage = df_covid.copy()

df_nursingstaffshortage.sort_values(by = ['Week Ending', 'Provider Name'])   #rearrange to make sure in date order and by name
df_nursingstaffshortageupdated = df_nursingstaffshortage.dropna(subset=['Shortage of Nursing Staff'])    #getting rid of nan values in Shortage of Nursing Staff column
#df_nursingstaffshortageupdated.shape                              checking shape

#new column calculating and storing num of consecutive weeks
df_nursingstaffshortageupdated['Consecutive Weeks of Shortage'] = (df_nursingstaffshortageupdated['Shortage of Nursing Staff'] == 'Y').groupby((df_nursingstaffshortageupdated["Shortage of Nursing Staff"] != df_nursingstaffshortageupdated["Shortage of Nursing Staff"].shift()).cumsum()).cumsum()
df_maxnursingstaffshortage = df_nursingstaffshortageupdated.groupby('Provider Name')['Consecutive Weeks of Shortage'].max()
df_maxnursingstaffshortage = df_maxnursingstaffshortage.sort_values(ascending = False)    #descending order

print(df_maxnursingstaffshortage.head(5))

```
